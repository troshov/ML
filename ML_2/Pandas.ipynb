Часть 3: Обработка DataFrame
Выбор данных
Pandas позволяет выбирать данные по индексам, меткам столбцов и условиям.

```python
df['Name']
```

```python
df[['Name', 'Age', 'Embarked']]
```

Сортировка данных
Сортировка данных по значениям столбцов.

```python
df.sort_values('PassengerId')
```

Группировка данных
Pandas также позволяет группировать данные и выполнять агрегирующие операции.

```python
surv_rate = df.groupby('Pclass')['Survived'].mean()
print(surv_rate)
```

Часть 4: Задания для практики
Прочитайте данные из файла 'titanic.csv'. Проверьте, есть ли пропущенные значения в данных и заполните их нулями. Выведите первые 10 строк данных. Выберите только те строки, где значение в столбце 'Age' больше 30. Отсортируйте данные по столбцу 'Fare' в порядке убывания. Сгруппируйте данные по столбцу 'Pclass' и вычислите средний возраст ('Age') для каждого класса.

```python
#Присваимаем переменной наш датасет, проверим вывод
df = pd.read_csv('Titanic.csv')
df
```

```python
#проверим на Nan
df.isna()
```

```python
#заполним пропуски нулями
df.fillna(0)
```

```python
#выведем первые 10 строк
df.head(10)
```

```python
#выберем строки, в которых Age > 30
df_age = df[df['Age'] > 30]
df_age
```

```python
#сделаем сортировку по убыванию по столбцу Fare
df_age.sort_values('Fare', ascending=False)
```

```python
#Распределим пассажиров по Pclass и определим их средний возраст
df_age.groupby('Pclass')['Age'].mean()
```
